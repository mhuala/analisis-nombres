```python
import pandas as pd
```

```python
df = pd.read_csv("Nombres 2010-2021.csv")
df.drop(columns=df.columns[0], axis=1, inplace=True)
df["year"] = df.year.astype('int32')
df["month"] = df.month.astype('int32')
df['date'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1))
df.head()
```

```python
df_grouped = df.groupby(["nombre","year"],as_index=False)["cantidad"].sum()
df_grouped_2 = df_grouped.sort_values(by=["nombre","year"])
df_grouped_2["diff"] = df_grouped_2["cantidad"].diff()
df_grouped_2.to_csv("df_final.csv",encoding="utf-8",index = False, sep = ";", decimal= ",")
```

```python
#df_grouped_2 = df_grouped_2.groupby(["nombre","year"],as_index=False)["cantidad"].diff().fillna(0)
df_grouped['pct_yoy'] = df_grouped['cantidad'].pct_change(periods=1)*100
df_grouped_mask = df_grouped.drop_duplicates("nombre",keep="first")
df_grouped_2_filtered =  pd.merge(df_grouped_2, df_grouped_mask, on=['nombre','year'], how='outer', indicator=True)
df_grouped_2_filtered = df_grouped_2_filtered[df_grouped_2_filtered["_merge"]=="left_only"]
df_grouped_2_filtered.to_csv("df_filtered.csv",encoding="utf-8",index = False, sep = ";", decimal= ",")
```
